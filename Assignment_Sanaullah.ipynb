5. Compare the lexical diversity scores for humor and romance fiction in 1.1. Which genre is more lexically diverse?

```python
#For this we can use a function or we can check it one by one before comparing

def lexical_diversity(text):
    return len(set(text)) / len(text)*100

lexical_diversity_H = lexical_diversity(text5)  # Assuming text5 is the Chat Corpus (humor)
lexical_diversity_R = lexical_diversity(text2)  # Assuming text2 is Sense and Sensibility (romance)

# Compare the scores
if lexical_diversity_H > lexical_diversity_R:
    print("Humor fiction is more lexically diverse.")
elif lexical_diversity_H < lexical_diversity_R:
    print("Romance fiction is more lexically diverse.")
else:
    print("Both genres have the same lexical diversity.")
```

22. Find all the four-letter words in the Chat Corpus (text5). With the help of a frequency distribution (FreqDist), show these words in decreasing order of frequency.

```python
four_letter_words = [word.lower() for word in text5 if len(word) == 4 and word.isalpha()]

freq_distribution = FreqDist(four_letter_words)

sorted_freq = sorted(freq_distribution.items(), key=lambda x: x[1], reverse=True)

for word, frequency in sorted_freq:
    print(f"{word}: {frequency}")
```
